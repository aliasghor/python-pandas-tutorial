# 2. Pemilihan baris (selection rows) pada `DataFrame`

```python
# Import all of the necessary packages
import numpy as np
import pandas as pd
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 5, size=(n_rows, n_cols)), columns=cols)
df
```

## Pemilihan baris menggunakan operator `|` (OR)

```python
df[(df["A"] == 4) | (df["A"] == 1)]
```

## Pemilihan baris menggunakan method `.isin()`

```python
df.isin?
```

```python
df[df["A"].isin([4, 1])]
```

```python
df.isin({"A": [4, 1]})
```

## Selection rows dengan menggunakan negasi (`~`)

```python
df[~df["A"].isin([4, 1])]
```
