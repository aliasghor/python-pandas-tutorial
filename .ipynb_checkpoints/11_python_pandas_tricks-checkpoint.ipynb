# 11. Melakukan seleksi kolom dan baris pada `DataFrame` menggunakan attribute `.loc`

```python
# Import all of the necessary packages
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 5
cols = list("ABCDE")

df1 = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)), columns=cols)
df1
```

```python
df2 = pd.read_csv("titanicfull.csv")
df2.head(10)
```

## Seleksi kolom dan baris menggunakan `.loc`

```python
df1.loc[[0, 3, 4], ["B", "E"]]
```

```python
df2.loc[[0, 3, 4], ["name", "sex", "age"]]
```

## Seleksi baris dengan kondisi

```python
df1.loc[df1.B > 10, ["B", "D", "E"]]
```

```python
df2.loc[df2.survived == 1, ["survived","name", "sex", "age"]].head(10)
```

## Slicing `DataFrame` dengan `.loc`

```python
df1.loc[1:4, "B":"D"]
```

```python
df2.loc[1:4, "survived":"age"]
```
