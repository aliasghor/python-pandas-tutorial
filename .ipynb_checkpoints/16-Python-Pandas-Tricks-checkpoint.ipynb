# 16. Menata ulang `DataFrame` dengan multiple indexes menggunakan `unstack()`

```python
# Import the Pandas package
import pandas as pd
```

## Persiapan `DataFrame`

```python
df = pd.read_csv("titanicfull.csv")
df.head(10)
```

## `DataFrame` dengan multiple indexes dari hasil groupping

```python
df.groupby(["sex", "pclass"])["survived"].mean().to_frame()
```

## Menata ulang `DataFrame` dengan multiple indexes

```python
df.groupby(["sex", "pclass"])["survived"].mean().unstack()
```
