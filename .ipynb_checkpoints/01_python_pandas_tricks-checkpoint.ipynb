# 1. Menyertakan Suffix dan Prefix pada kolom `DataFrame`

```python
# Import all of the necessary packages
import numpy as np
import pandas as pd
```

## Persiapan `DataFrame`

```python
n_rows = 5
n_cols = 5
cols = tuple("ABCDE")

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)), columns=cols)
df
```

## Menyertakan Suffix

```python
df.add_suffix?
```

```python
df.add_suffix("_col")
```

## Menyertakan Prefix

```python
df.add_prefix?
```

```python
df.add_prefix("col_")
```
