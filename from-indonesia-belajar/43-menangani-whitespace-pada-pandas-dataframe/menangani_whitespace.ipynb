# 43. Menangani whitespace pada `DataFrame`

# Import modules

```python
import pandas as pd
import numpy as np
```

# Persiapan `DataFrame`

```python
data = {
    "nim":["10","11","12","13"," "],
    "nama":["Adi"," ","Tejo"," ","Bejo"]
}

df = pd.DataFrame(data)
df
```

# Menangani whitespace pada `DataFrame`

```python
df.info()
```

```python
df = df.replace(r"^\s*$",np.nan,regex=True)
df
```
