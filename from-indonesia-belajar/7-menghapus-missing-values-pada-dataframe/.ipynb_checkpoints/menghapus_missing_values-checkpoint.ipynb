# 7. Menghapus missing values pada `DataFrame`

## Import packages

```python
import pandas as pd
```

```python
df = pd.read_csv('../titanicfull.csv')
df.head()
```

```python
df_backup = df.copy(deep=True)
```

## Menghapus (drop) setiap kolom yang mengandung missing values (`NaN`)

```python
df = df.dropna(axis='columns')
df.head()
```

## Menghapus (drop) setiap rows yang mengandung missing values (`NaN`)

```python
df = df_backup

df = df.dropna(axis='rows')
df.head()
```

## Melihat presentase missng values (`NaN`) untuk setiap kolom

```python
df = df_backup

df.isna().mean()
```

## Menghapus (drop) missing values (`NaN`) untuk setiap kolom berdasarkan threshold

```python
treshold = len(df) * 0.9
df = df.dropna(thresh=treshold,axis='columns')
df.head()
```
