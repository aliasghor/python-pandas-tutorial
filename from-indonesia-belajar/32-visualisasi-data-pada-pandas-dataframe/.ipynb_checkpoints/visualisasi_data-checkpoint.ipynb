# 32. Plotting dari suatu Pandas `DataFrame`

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 40
n_cols = 5
cols = list("ABCDE")

df = pd.DataFrame(np.random.randint(1,20,size=(n_rows,n_cols)),columns=cols)
df.head()
```

## Line Plot

```python
df.plot(kind="line")
```

```python
df[["A","B"]].plot(kind="line")
```

## Bar Plot

```python
df.plot(kind="bar")
```

```python
df[["A","B"]].plot(kind="bar")
```

```python
df[["A","B"]].head().plot(kind="bar")
```

```python
df[["A","B"]].head().plot(kind="barh")
```

## Area Plot

```python
df.plot(kind="area")
```

```python
df[["A","B"]].head().plot(kind="area")
```

## Box Plot

```python
df.plot(kind="box")
```

## Histogram

df.plot(kind="hist")

```python
df[["A","B"]].plot(kind="hist")
```

# Kernel Density Estimation (KDE)

```python
df.plot(kind="kde")
```

## Scatter Plot

```python
df.plot(x="A",y="B",kind="scatter")
```
