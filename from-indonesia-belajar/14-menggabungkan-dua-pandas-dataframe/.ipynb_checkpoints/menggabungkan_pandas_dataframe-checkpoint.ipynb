# 14. Menggabungkan dua Pandas `DataFrame`

```python
import pandas as pd
import numpy as np
```

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,5,size=(n_rows,n_cols)),columns=cols)
df
```

```python
df1 = df.copy(deep=True)
df1 = df1.drop([1,4])
df1
```

```python
df2 = df.copy(deep=True)
df2 = df2.drop([0,3])
df2
```

## Menggabungkan dua `DataFrame`

```python
df_inner = pd.merge(df1,df2,how='inner')
df_inner
```

```python
df_outer = pd.merge(df1,df2,how='outer')
df_outer
```
