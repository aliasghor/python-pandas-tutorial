# 34. Mapping pada `DataFrame` menggunakan `applymap()`

## Import module

```python
import pandas as pd

pd.__version__
```

## Persiapan `DataFrame`

```python
data = {
    "jenis_kelamin":["pria","Wanita","lelaki","Lelaki","perempuan"],
    "usia":[21,22,23,24,25],
    "shift":["pagi","Siang","Malam","siang","pagi"]
}

df = pd.DataFrame(data)
df
```

## Mapping pada `DataFrame` dengan `applymap()`

```python
df = df.applymap(lambda x: x.lower() if type(x) == str else x)
df
```

## `applymap()` dengan dictionary

```python
mapping = {
    "pria":"L",
    "Lelaki":"L",
    "lelaki":"L",
    "wanita":"P",
    "perempuan":"P",
    "pagi":1,
    "siang":2,
    "malam":3
}

df.applymap(mapping.get)
```

```python
df[["jenis_kelamin","shift"]] = df[["jenis_kelamin","shift"]].applymap(mapping.get)
df
```
