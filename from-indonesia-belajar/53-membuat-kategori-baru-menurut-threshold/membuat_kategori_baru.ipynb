# 53. Membuat kategori baru berdasarkan threshold (ambang batas)

# Import module

```python
import pandas as pd
```

# Persiapan `DataFrame`

```python
data = {
    "hobi":["jogging","mancing","berenang","mancing","mancing","membaca","membaca","mancing","fotografi","mancing","camping"]
}

df = pd.DataFrame(data)
df
```

# Membuat kategori baru berdasarkan threshold (ambang batas)

```python
df["hobi"].value_counts()
```

```python
persentase = df["hobi"].value_counts(normalize=True)
persentase
```

```python
threshold = 0.1
lainnya = persentase[persentase < threshold].index
lainnya
```

```python
df["hobi"] = df["hobi"].replace(lainnya,"lainnya")
df
```
