#  20. Menggabungkan (merge) dua `DataFrame` secara berdampingan

## Import module

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

## Persiapan `DataFrame`

```python
n_rows = 5
n_cols = 2
cols1 = list("AB")

df1 = pd.DataFrame(np.random.randint(1,5,size=(n_rows,n_cols)),columns=cols1)
df1
```

```python
cols2 = list("CD")

df2 = pd.DataFrame(np.random.randint(1,10,size=(n_rows,n_cols)),columns=cols2)
df2
```

## Menggabungkan (merge) dua `DataFrame` secara berdampingan

```python
df = pd.merge(df1,df2,left_index=True,right_index=True)
df
```
