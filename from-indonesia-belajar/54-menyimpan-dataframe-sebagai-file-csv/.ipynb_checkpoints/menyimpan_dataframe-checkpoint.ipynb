# 54. Menyimpan `DataFrame` sebagai file `CSV`

# Import module

```python
import pandas as pd
```

# Persiapan `DataFrame`

```python
data = {
    "Nama":["Bayu","Indra","Devi","Agni"],
    "Jenis_kelamin":["L","L","P","L"],
    "Usia":[23,21,22,25]
}

df = pd.DataFrame(data)
df
```

# Menyimpan `DataFrame` sebagai file `csv`

```python
df.to_csv("data_peserta1.csv")
```

```python
df = pd.read_csv("data_peserta1.csv")
df
```

```python
df = pd.read_csv("data_peserta1.csv",index_col=0)
df
```

```python
df.to_csv("data_peserta2.csv",index=False)
```
