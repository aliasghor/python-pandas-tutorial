# 8. Memeriksa kesamaan antar kolom (Series) pada `DataFrame`

## Import packages

```python
import pandas as pd
import numpy as np
```

```python
data = {
    'A':[1,2,3,np.nan,5],
    'B':[1,2,3,np.nan,5]
}

df = pd.DataFrame(data)
df
```

## Apa itu `Series`?
`Series` adalah sebuah kolom tunggal dari `DataFrame`.

```python
# Membuat object Series

series = pd.Series(["Ali","Mogi","Sampi","Balu"],name='Name')
series
```

```python
type(series)
```

## Memerika kesamaan `Series` dengan operator `==`

```python
df['A'] == df['B']
```

## Memeriksa kesamaan `Series` menggunakan `.equals()` method

```python
df['A'].equals(df['B'])
```

## Memeriksa antar dua buah kolom `DataFrame` menggunakan `.equals()` method

```python
another_df = df.copy(deep=True)

another_df.equals(df)
```
