# 30. Akses nilai variabel dalam fungsi `query`

## Import module

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 5
cols = list("ABCDE")

df = pd.DataFrame(np.random.randint(1,20,size=(n_rows,n_cols)),columns=cols)
df
```

## Akses nilai variabel pada `query()`

```python
df.query('A < 5')
```

```python
df.query("A < 5 and B > 10")
```

```python
rerata = df["A"].mean()
rerata
```

```python
df.query("A > @rerata")
```
