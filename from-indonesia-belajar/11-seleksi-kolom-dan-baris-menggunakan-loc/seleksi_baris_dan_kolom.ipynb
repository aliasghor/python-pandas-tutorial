# 11. Seleksi kolom dan baris pada `DataFrame` menggunakan attribute `.loc`

## Import pandas package

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
df = pd.read_csv('../titanicfull.csv')
df.head()
```

```python
df_backup = df.copy(deep=True)
```

## Seleksi kolom dan baris menggunakan attribute `.loc`

```python
df.loc[[0,3,4],['name','sex','age']]
```

## Seleksi baris dengan kondisi

```python
df = df.loc[df['age'] < 18,['name','sex','age']]
df.head()
```

## Slicing `DataFrame` menggunakan attribute `.loc`

```python
df = df_backup

df.loc[0:5,'survived':'age']
```
