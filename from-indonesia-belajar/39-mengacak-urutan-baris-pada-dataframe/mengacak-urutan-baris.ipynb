# 39. Mengacak urutan baris pada `DataFrame`

## Import modules

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 6
n_cols = 5
cols = list("ABCDE")

df = pd.DataFrame(np.random.randint(1,5,size=(n_rows,n_cols)),columns=cols)
df
```

## Mengacak baris pada `DataFrame`

```python
df.sample(frac=1.0,random_state=1)
```

```python
df.sample(frac=1.0,random_state=1,ignore_index=True)
```
