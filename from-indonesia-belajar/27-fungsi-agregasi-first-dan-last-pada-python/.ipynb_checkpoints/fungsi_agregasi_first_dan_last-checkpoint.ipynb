# 27. Fungsi agregasi first dan last pada Python Pandas

## Import module

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
data = {
    "Dokter":["Budi","Wati","Iwan","Budi","Budi","Wati"],
    "Pasien":["Abdul","Rahmat","Asep","Joko","Wiwin","Lisa"]
}

df = pd.DataFrame(data)
df
```

## Mengenal fungsi agregasi `first()` dan `last()` 

```python
# Menghitung jumlah total pasien untuk setiap masing-masing dokter
df.groupby("Dokter")["Pasien"].count().to_frame()
```

```python
# Menampilkan nama pasien pertama untuk masing-masing dokter
df.groupby("Dokter")["Pasien"].first().to_frame()
```

```python
# Menampilkan nama pasien terakhir untuk masing-masing dokter
df.groupby("Dokter")["Pasien"].last().to_frame()
```
