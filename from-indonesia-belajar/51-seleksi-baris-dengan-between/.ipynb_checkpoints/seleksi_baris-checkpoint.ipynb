# 51. Seleksi baris dengan `between()`

# Import modules

```python
import pandas as pd
import numpy as np
```

# Persiapan `DataFrame`

```python
cols = 5
rows = 10
columns = list("ABCDE")

df = pd.DataFrame(np.random.randint(1,10,size=(rows,cols)),columns=columns)
df
```

# Seleksi baris dengan `between()`

```python
df[df["A"].between(2,5)]
```

```python
df[df["B"].between(2,5)]
```
