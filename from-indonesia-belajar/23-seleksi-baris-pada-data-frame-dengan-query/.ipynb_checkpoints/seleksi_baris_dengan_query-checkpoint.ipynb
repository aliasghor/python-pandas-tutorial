# 23. Seleksi baris pada `DataFrame` dengan query

## Import module

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
# Manually create a DataFrame object from a dictionary consits of a lists
n_rows = 10
n_cols = 2
cols = ["kolom_satu","kolom dua"]

df1 = pd.DataFrame(np.random.randint(1,20,size=(n_rows,n_cols)),columns=cols)
df1
```

```python
# Create a backup DataFrame object
df_backup = df1.copy(deep=True)
```

```python
# Read titanicfull.csv file and create a DataFrame
df2 = pd.read_csv("../titanicfull.csv")
df2.head()
```

## Seleksi baris dengan `.query()`

```python
df2.query("age < 18")
```

```python
# Selecting a column without underscore using .query() method
# First technique
df1.query("`kolom dua` > 18")
```

```python
# Second technique
df1 = df1.rename(columns={"kolom dua":"kolom_dua"})
df1.query('kolom_dua > 18')
```
