# 18. Membentuk dummy `DataFrame` Python Pandas

```python
import pandas as pd
import numpy as np
```

## Membentuk object `DataFrame` dari dictionary beserta dengan sekumpulan lists pada dictionary

```python
data = {
    'Name':['Ali','Mogi','Gerry'],
    'Age':[21,18,17],
    'Gender':['Male','Female','Male']
}

df = pd.DataFrame(data)
df
```

## Membentuk object `DataFrame` dari `NumPy` array beserta dengan missing values (`NaN`)

```python
n_rows = 10
n_cols = 5
cols = list('ABCDE')
rng = np.random.default_rng()

df = pd.DataFrame(rng.integers(1,10,size=(n_rows,n_cols)),columns=cols)
df.loc[0:5,'A'] = np.nan
df.loc[5:6,'E'] = np.nan

df
```
