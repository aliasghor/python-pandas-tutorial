# 35. Memadukan fungsi agregasi dengan `transform()`

## Import module

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
data = {
    "no_nota":[1,1,1,2,2,3,4,5],
    "kopi":["latte","cappuccino","espresso","latte","espresso","cappuccino","latte","espresso"],
    "harga":[50,60,80,150,120,60,100,40]
}

df = pd.DataFrame(data)
df
```

## Menghitung total harga untuk setiap nota

```python
df.groupby("no_nota")["harga"].sum().to_frame()
```

```python
df["total_harga"] = df.groupby("no_nota")["harga"].transform(sum)
df
```

## Menghitung total omset untuk setiap kopi yang terjual

```python
df.groupby("kopi")["harga"].sum().to_frame()
```

```python
df["total_omset"] = df.groupby("kopi")["harga"].transform(sum)
df
```

```python
df[["kopi","total_omset"]].plot(kind="bar",ylabel="Total omset",xlabel="Nama kopi berdasarkan index DataFrame",title="Laporan omset penjualan kopi")
```
