# 15. Memecah nilai string suatu kolom ke dalam beberapa kolom baru pada Pandas `DataFrame`

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
data = {
    'President\'s_name':['Bung Karno','Soeharto','Bacharuddin Jusuf Habibi','Susilo Bambang Yudhoyono','Joko Widodo'],
    'Born_Place':['Surabaya, Jawa Timur','Bantul, Yogyakarta','Bugis, Gorontalo','Pacitan, Jawa Timur','Surakarta, Jawa Tengah']
}

df = pd.DataFrame(data)
df
```

## Memecah nama Kota dan Provinsi

```python
df[['City','Province']] = df['Born_Place'].str.split(',',expand=True)
df
```
