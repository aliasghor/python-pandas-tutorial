# 28. Mengenal `explode` dan `implode` List pada Pandas `DataFrame`

## Import module

```python
import pandas as pd

pd.__version__
```

```python
data = {
    "Team":["Dc","Marvel"],
    "Heroes":[["Batman","Superman","Wonder Woman","Aquaman","Green Lantren","Shazam"],["Iron Man","Captain America","Ant-Man","Black Panther","Captain Marvel"]]
}

df = pd.DataFrame(data)
df
```

## Explode

```python
df.explode?
```

```python
df1 = df.explode('Heroes')
df1
```

```python
# Menggunakan parameter `ignore_index=True`
df2 = df.explode("Heroes",ignore_index=True)
df2
```

## Implode

```python
d = {
    "Team":["Dc","Marvel"]
}

df3 = pd.DataFrame(d)
df3
```

```python
df3["Imploded"] = df1.groupby(df1.index)["Heroes"].agg(list)
df3
```
