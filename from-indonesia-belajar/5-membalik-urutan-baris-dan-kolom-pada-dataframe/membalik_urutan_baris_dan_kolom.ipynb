# 5. Membalik urutan baris dan kolom pada `DataFrame`

## Import packages

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,5,size=(n_rows,n_cols)),columns=cols)
df
```

## Membalik urutan kolom

```python
df.loc[:,::-1]
```

## Membalik urutan baris

```python
df.loc[::-1]
```

# Membalik urutan baris dan melaukan penyesuaian index

```python
df.loc[::-1].reset_index(drop=True)
```
