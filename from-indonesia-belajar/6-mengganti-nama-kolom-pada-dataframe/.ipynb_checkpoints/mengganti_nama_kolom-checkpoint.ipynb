# 6. Mengganti nama kolom pada `DataFrame`

## Import packages

```python
import pandas as pd
```

```python
data = {
    'A':['Ali','Gerry','Mogi'],
    'B':[21,20,19]
}

df = pd.DataFrame(data)
df
```

## Mengganti nama (label) untuk sebuah kolom pada `DataFrame`

```python
df = df.rename(columns={'A':'Name'})
df
```

```python
df = df.rename(columns={'B':'Age'})
df
```
