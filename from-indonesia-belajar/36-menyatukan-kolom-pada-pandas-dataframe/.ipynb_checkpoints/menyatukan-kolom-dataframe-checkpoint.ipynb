# Meyatukan kolom dengan `str.cat()`

## Import module

```python
import pandas as pd

pd.__version__
```

## Persiapan `DataFrame`

```python
data = {
    "Nama":["Ali","Bayu","Indra","Devi"],
    "Jenis_kelamin":["L","L","L","P"],
    "Usia":[21,23,22,25]
}

df = pd.DataFrame(data)
df
```

## Menyatukan kolom dengan `str.cat()`

```python
df["Nama"].str.cat(df["Jenis_kelamin"],sep=", ").to_frame()
```

```python
df["Nama_dan_Jenis_kelamin"] = df["Nama"].str.cat(df["Jenis_kelamin"],sep=", ")
df
```

```python
df["Nama"].str.cat(df["Usia"].astype("str"), sep=" - ").to_frame()
```

```python
df["Nama_dan_Usia"] = df["Nama"].str.cat(df["Usia"].astype("str"), sep=" - ")
df
```
