# 37. Seleksi baris dengan banyak kriteria

## Import module

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
df = pd.read_csv("../titanicfull.csv")
df.head()
```

## Seleksi baris dengan banyak kriteria

```python
# Mengubah tipe data dari kolom survived menjadi boolean
df["survived"] = df["survived"].astype("bool")
df.head()
```

```python
# Contoh pertama dari seleksi baris dengan banyak kriteria
df[(df["sex"] == "female") & (df["survived"] == True) & (df["embarked"] == "S") & (df["age"] >= 60)]
```

```python
# Contoh kedua dari seleksi baris dengan banyak kriteria
df[
    (df["sex"] == "female") &
    (df["age"] >= 60) &
    (df["survived"] == True) &
    (df["embarked"] == "S")
]
```

```python
# Contoh ketiga dari seleksi bari dengan banyak kriteria
kr1 = df["sex"] == "female"
kr2 = df["survived"] == True
kr3 = df["embarked"] == "S"
kr4 = df["age"] >= 60

df[kr1 & kr2 & kr3 & kr4]
```
