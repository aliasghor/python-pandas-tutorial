# 16. Menata ulang `DataFrame` dengan multiple indexes menggunakan `unstack()`

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
df = pd.read_csv('../titanicfull.csv')
df.head()
```

```python
df['survived'] = df['survived'].astype('bool')
df.head()
```

## `DataFrame` dengan multiple indexes dari hasil groupping

```python
df.groupby(['sex','pclass'])['survived'].mean().to_frame()
```

## Menata ulang `DataFrame` dengan multiple indexes

```python
df.groupby(['sex','pclass'])['survived'].mean().unstack()
```
