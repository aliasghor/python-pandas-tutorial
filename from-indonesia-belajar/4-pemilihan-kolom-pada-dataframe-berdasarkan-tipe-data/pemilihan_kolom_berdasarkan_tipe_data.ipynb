# 4. Pemilihan kolom (columns selection) pada Pandas `DataFrame` berdasarkan tipe data

## Import packages

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 5
n_cols = 4
cols = ['bilangan_bulat','bilangan_pecahan','waktu','teks']

df = pd.DataFrame(np.random.randint(1,20,size=(n_rows,n_cols)),columns=cols)

df['waktu'] = pd.date_range('1/1/2000',periods=n_rows)
df['bilangan_pecahan'] = np.random.rand(n_rows)
df['teks'] = tuple('ABCDE')

df
```

```python
df.dtypes
```

## Memilih kolom bertipe data numerik

```python
df.select_dtypes(include='number')
```

```python
df.select_dtypes(include='float')
```

```python
df.select_dtypes(include='int')
```

## Memilih kolom bertipe data datetime

```python
df.select_dtypes(include='datetime')
```

## Memilih kolom bertipe data string atau `object`

```python
df.select_dtypes(include='object')
```

## Memilih kolom dengan kombinasi tipe data

```python
df.select_dtypes(include=['datetime','object'])
```
