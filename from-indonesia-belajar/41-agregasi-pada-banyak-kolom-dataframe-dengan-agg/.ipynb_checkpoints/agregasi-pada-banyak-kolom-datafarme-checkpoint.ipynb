# 41. Menerapkan agregasi pada sejumlah kolom dengan `agg()`

## Import module

```python
import pandas as pd

pd.__version__
```

## Persiapan `DataFrame`

```python
df = pd.read_csv("../titanicfull.csv")
df.head()
```

```python
df["survived"] = df["survived"].astype("bool")
df.head()
```

## Menerapkan agregasi pada sejumlah kolom dengan `agg()`

```python
df.groupby("pclass").agg({
    "pclass":"count",
    "age":["mean","max","min"],
    "survived":"mean"
})
```

```python
df.groupby("pclass").agg(n_pass=("pclass","count"),avg_age=("age","mean"),max_age=("age","max"),min_age=("age","min"),survival_rate=("survived","mean"))
```
