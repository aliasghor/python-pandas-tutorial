# 1. Menyertakan Prefix dan Suffix pada seluruh Kolom Pandas `DataFrame`

## Import pacakges

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 5
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10,size=(n_rows,n_cols)),columns=cols)
df
```

## Menyertakan prefix (string bagian depan) ke keseluruhan kolom `DataFrame`

```python
df.add_prefix('Kolom_')
```

## Menyertakan suffix (string bagian belakang) ke keseluruhan kolom `DataFrame`

```python
df.add_suffix('_field')
```
