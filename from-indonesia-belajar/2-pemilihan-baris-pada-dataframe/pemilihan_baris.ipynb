# 2. Pemilihan baris (rows selection) pada `DataFrame`

## Import packages

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 5
cols = tuple('ABCDE')

df = pd.DataFrame(np.random.randint(1,10,size=(n_rows,n_cols)),columns=cols)
df
```

## Rows selection menggunakan operator logika `|`

```python
df[(df['A'] == 7) | (df['A'] == 8)]
```

## Rows selection menggunakan method `.isin()`

```python
df[df['A'].isin([7,8])]
```

## Rows selection menggunakan negasi `~`

```python
df[~df['A'].isin([7,8])]
```
