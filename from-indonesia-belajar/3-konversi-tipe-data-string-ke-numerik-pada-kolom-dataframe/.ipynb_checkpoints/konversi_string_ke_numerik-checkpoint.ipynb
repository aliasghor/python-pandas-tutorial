# 3. Konversi tipe data String ke Numerik pada kolom `DataFrame`

## Import packages

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
data = {
    'Kolom_A':['1','2','3','4','5'],
    'Kolom_B':['1','2','3','4','Ali Ganteng']
}

df = pd.DataFrame(data)
df
```

```python
df.dtypes
```

## Konversi tipe data dengan menggunakan method `.astype()`

```python
df = df.astype({'Kolom_A':'int'})
df
```

```python
df.dtypes
```

## Konversi keseluruhan kolom menjadi numerik

```python
df = df.apply(pd.to_numeric,errors='coerce')
df
```
