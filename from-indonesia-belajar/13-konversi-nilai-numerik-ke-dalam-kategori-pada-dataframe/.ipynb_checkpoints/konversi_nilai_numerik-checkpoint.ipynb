# 13. Konversi nilai numerik ke dalam kategori pada `DataFrame`

```python
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 1
cols = ('usia',)

df = pd.DataFrame(np.random.randint(1,70,size=(n_rows,n_cols)),columns=cols)
df
```

## Pengelompokan nilai numerik ke dalam beberapa kategori menggunakan `cut()`

```python
df['kelompok_usia'] = pd.cut(df['usia'],bins=[0,18,40,70],labels=['anak','dewasa','manula'])
df
```
