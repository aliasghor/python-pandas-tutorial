# 31. Tipe data ordinal pada pandas `DataFrame`

## Import module

```python
import pandas as pd
```

## Persiapan `DataFrame`

```python
data = {
    "Nama":["Ali","Gerry","Mogi","Sampi"],
    "Tingkat_kepuasaan":["Sangat memuaskaan","Memuaskaan","Cukup","Tidak memuaskaan"]
}

df = pd.DataFrame(data)
df
```

## Tipe data ordinal pada pandas

```python
from pandas.api.types import CategoricalDtype

tingkat_kepuasaan = CategoricalDtype(["Tidak memuaskaan","Cukup","Memuaskaan","Sangat memuaskaan"],ordered=True)

df["Tingkat_kepuasaan"] = df["Tingkat_kepuasaan"].astype(tingkat_kepuasaan)
df
```

```python
df.sort_values(by="Tingkat_kepuasaan",ignore_index=True)
```

```python
df[df["Tingkat_kepuasaan"] > "Cukup"]
```
