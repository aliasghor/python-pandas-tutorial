# 6. Mengganti nama (label) pada kolom `DataFrame`

```python
# Import the necessary package
import pandas as pd
```

## Persiapan `DataFrame`

```python
data = {
    "A": ["Gerry", "Ali", "Sampi", "Balu", "Apooh"],
    "B": ["Basket", "Coding", "Berenang", "Badminton", "Membaca buku"],
    "C": [30, 22, 20, 19, 40]
}

df = pd.DataFrame(data)
df
```

## Mengganti nama (label) untuk sebuah kolom `DataFrame`

```python
df.rename?
```

```python
df = df.rename(columns={"B": "Hobby"})
df
```

## Mengganti nama (label) untuk banyak kolom `DataFrame`

```python
df = df.rename(columns={"A": "Name", "C": "Age"})
df
```
