# 13. Konversi nilai numerik kedalam sejumlah kategori

```python
# Import all of the necessary packages
import pandas as pd
import numpy as np
```

## Persiapan `DataFrame`

```python
n_rows = 10
n_cols = 1
cols = ("Usia",)

df = pd.DataFrame(np.random.randint(1, 99, size=(n_rows, n_cols)), columns=cols)
df
```

## Pengelompokan nilai numerik kedalam beberapa kategori menggunakan `cut()`

```python
pd.cut?
```

```python
df["Kelompok_Usia"] = pd.cut(df["Usia"], bins=[0, 15, 18, 40, 100], labels=["Anak", "Remaja", "Dewasa", "Manula"])
df
```
